```python
df_original=pd.read_excel('Bollywood Dataset.xlsx')
df1=df_original.copy()
```

```python
# Top 10 Movies by Popularity


top_movies = df1.sort_values(by='popularity', ascending=False).head(10)

# Create the plot
plt.figure(figsize=(10,6))
bar_plot = sns.barplot(x=top_movies.iloc[:, 10], y=top_movies['popularity'])  # Adjust the column index if needed
plt.xlabel('Movie Title')  # Change this to the correct column name for better clarity
plt.ylabel('Popularity of movie')
plt.title('Top 10 Movies by Popularity')
plt.xticks(rotation=45)  # Rotate x labels for better visibility

for bar in bar_plot.containers:
    bar_plot.bar_label(bar)
    
plt.show()

```

```python

# Assuming 'director' is the name of the column with director names, and 'popularity' is the column with popularity scores
top_movie = df1.sort_values(by='popularity', ascending=False).head(10)

plt.figure(figsize=(10, 6))
bar_plot=sns.barplot(x=top_movie['Director'], y=top_movie['popularity'])
plt.xticks(rotation=45, ha='right')  # Rotate director names for better visibility
plt.xlabel('Director')
plt.ylabel('Popularity of movie')
plt.title('Top 10 Popularity of movies by Director')
plt.tight_layout()  # Adjust layout to fit everything properly

for i in bar_plot.containers:
    bar_plot.bar_label(i)

plt.show()
```

```python
# Assuming 'director' is the name of the column with director names, and 'popularity' is the column with popularity scores
top_movie = df1.sort_values(by='popularity', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_movie['Actor_Name'], y=top_movie['popularity'])
plt.xticks(rotation=45, ha='right')  # Rotate director names for better visibility
plt.xlabel('Actor')
plt.ylabel('Popularity of movie')
plt.title('Top 10 Popularity of movies by Actor')
plt.tight_layout()  # Adjust layout to fit everything properly
plt.show()
```

```python
# Convert 'release_date' column to datetime format
df1['release_date'] = pd.to_datetime(df1['release_date'])

# Step 2: Extract the year from the 'release_date' column
df1['release_year'] = df1['release_date'].dt.year

```

```python
# count which year most movie released



year_counts = df1['release_year'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
count_plot = sns.countplot(x='release_year', data=df1, order=year_counts.index)
plt.title('Number of movie release per year ')
# Add labels to the bars
for i in count_plot.containers:
    count_plot.bar_label(i)

plt.show()
```

```python
# Which Actor release most movies

Actors_count=df1['Actor_Name'].value_counts().head(10)
plt.figure(figsize=(12,6))
plt.xticks(rotation=45,ha='right')
plt.ylabel('Count of Movie released')
plt.xlabel('Actors')
Count_of_movie_released=sns.barplot(x=Actors_count.index,y=Actors_count.values)
for i in Count_of_movie_released.containers:
    Count_of_movie_released.bar_label(i)

plt.show()
```
